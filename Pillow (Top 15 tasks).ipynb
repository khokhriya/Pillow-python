## Python Image Library(PIL)

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1- Reading

```python
image= Image.open("ronaldo.jpg")
image2= Image.open("messy.jpg")
```

### 2- Displaying

```python
plt.imshow(image)
```

```python
plt.imshow(image2)
```

### 3-Displaying Properties of Image

```python
print(image.size)
print(image.format)
print(image.mode)
```

### 4- Saving

```python
image3=image.save("ronaldo_save.jpg")
```

### 5-Croping

```python
left=20.0
top=150.0
right=200.0
bottom=100.0
crop_ronaldo=image.crop((left,top,right,bottom))
```

```python
plt.imshow(crop_ronaldo)
```

### 6- Copying

```python
copied_image=image.copy()
plt.imshow(copied_image)
```

### 7- Transposing

```python
transpose_image1=image.transpose(Image.FLIP_LEFT_RIGHT)
transpose_image2=image.transpose(Image.FLIP_TOP_BOTTOM)
transpose_image3=image.transpose(Image.ROTATE_90)
transpose_image4=image.transpose(Image.ROTATE_180)
plt.figure(figsize=(10,10))

plt.subplot(3,2,1)
plt.imshow(transpose_image1)
plt.title("FLIP_LEFT_RIGHT")

plt.subplot(3,2,2)
plt.imshow(transpose_image2)
plt.title("FLIP_TOP_BOTTOM")

plt.subplot(3,2,3)
plt.imshow(transpose_image3)
plt.title("ROTATE_90")

plt.subplot(3,2,4)
plt.imshow(transpose_image4)
plt.title("ROTATE_180")
```
