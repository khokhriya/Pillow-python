```python
# Importing all required Libraries
import PIL
from PIL import Image
from IPython.display import display
file= "ronaldo.jpg"
image=Image.open(file)
```

```python
#Copy Image
help(image.copy)
```

```python
# Image save
help(image.save)
```

```python
#save and Inspect Image
image.save("ronaldo.png")
image=Image.open("ronaldo.png")
import inspect
inspect.getmro(type(image))
```

```python
# Import ImageFilter
from PIL import ImageFilter
help(ImageFilter)
```

```python
#Converting Image to "RGB" & Blurr image
image=image.convert('RGB') #RGB=red,green,blue
blurred_image= image.filter(PIL.ImageFilter.BLUR)
display(blurred_image)
```

```python
#print image size and shape
print("{}x{}".format(image.width, image.height))
```

```python
#image croping
#help(image.crop)
display(image.crop((100,20,200,120)))
```

```python
# ImageDraw
from PIL import ImageDraw
drawing_object=ImageDraw.Draw(image)
drawing_object.rectangle((50,0,190,150),fill=None ,outline='green')
display(image)
```

## Additional PILLOW functions

```python
# Image Enhancing
# PIL.Image.new function
# contact sheet
# grid system
```

```python
# Image Enhancing
from PIL import ImageEnhance
enhancer= ImageEnhance.Brightness(image)
images=[]
for i in range(0,10):
    images.append(enhancer.enhance(i/10))
print(images)
```

```python
# PIL.Image.new function
help(PIL.Image.new)
```

```python
first_image=images[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width,10*first_image.height))

# creating Counter variable for y location
current_location = 0
for img in images:
    # Lets paste the current image into the contact sheet
    contact_sheet.paste(img, (0, current_location) )
    # And update the current_location counter
    current_location=current_location+450
    
#contact sheet
contact_sheet = contact_sheet.resize((160,900) )

display(contact_sheet)
```

```python
#resize contact sheet in grid of 3*3
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,first_image.height*3))

#creating variables for x,y dimensions
x=0
y=0

for img in images[1:]:
    #paste image
    contact_sheet.paste(img,(x,y))
    
    if x+first_image.width == contact_sheet.width:
        x=0
        y=y+first_image.height
    else:
        x=x+first_image.width
# converting into int()        
contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))

display(contact_sheet)
```
